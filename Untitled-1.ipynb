```python
#VGG网络构建
import tensorflow as tf 
#模型构建
def vgg_block(num_conv,num_filters):
    #构建训练模型
    blk = tf.keras.models.Sequential()
    #遍历卷积层
    for _ in range(num_conv):
        #设置卷积层
        blk.add(tf.keras.layers.Conv2D(num_filters,kernel_size=3,padding="same",activation="relu"))
    #池化层
    blk.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
    return blk 
```

```python
#构建模型
def vgg(conv_arch):
    #序列模型
    net = tf.keras.models.Sequential()
    #生成卷积部分
    for (num_conv,num_filters) in conv_arch:
        net.add(vgg_block(num_conv,num_filters))
    #全连接层
    net.add(tf.keras.models.Sequential([
        #战平
        tf.keras.layers.Flatten(),
        #全连接层
        tf.keras.layers.Dense(4096,activation="relu"),
        #随机失火
        tf.keras.layers.Dropout(0.5),
        #输出层
        tf.keras.layers.Dense(10,activation="softmax")
    ]))
    return net 
```

```python
#lnception模块
class Inception(tf.keras.layers.Layer):
    #设置模块的构成
    def __init__(self,c1,c2,c3,c4):
        super().__init__()
        #线路 1*1 relu，same  c1
        self.p1_1 = tf.keras.layers.Conv2D(c1,kernel_size=1,activation="relu",padding="same")
        #线路2 1*1 relu same  c2【0】
        self.p2_1 = tf.keras.layers.Conv2D(c2[0],kernel_size=1,activation="relu",padding="same")
        #线路2 3*3 relu same c2[1]
        self.p2_2 = tf.keras.layers.Conv2D(c2[1],kernel_size=3,activation="relu",padding="same")
        #线路3 1*1 relu same  c2【0】
        self.p3_1 = tf.keras.layers.Conv2D(c3[0],kernel_size=1,activation="relu",padding="same")
        #线路2 5*5 relu same c2[1]
        self.p3_2 = tf.keras.layers.Conv2D(c3[1],kernel_size=5,activation="relu",padding="same")
        #线路4 max-pool 1*1
        self.p4_1 = tf.keras.layers.MaxPool2D(pool_size=3,padding="same",strides=1)
        #线路4  1*1卷积
        self.p4_2 = tf.keras.layers.Conv2D(c4,kernel_size=1,activation="relu",padding="same")
    #前行传播过程
    def call(self,x):
        #线路1 
        p1 = self.p1_1(x)
        #线路2
        p2 = self.p2_2(self.p2_1(x))
        #线路三
        p3 = self.p3_2(self.p3_1(x))
        #线路4
        p4 = self.p4_2(self.p4_1(x))
        #concat
        outputs = tf.concat([p1,p2,p3,p4],axis=-1)
        return outputs
```

```python
#辅助分类器
def aux_classifier(x,filter_size):
    #池化层
    x = tf.keras.layers.AveragePooling2D(pool_size=5,strides=3,padding="same")(x)
    #卷积层
    x = tf.keras.layers.Conv2D(filters=filter_size[0],kernel_size=1,strides=1,padding="valid",activation="relu")(x)
    #展平
    x = tf.keras.layers.Flatten()(x)
    #全连接
    x = tf.keras.layers.Dense(units=filter_size[1],activation="relu")(x)
    #输出层
    x = tf.keras.layers.Dense(units=10,activation="softmax")(x)
    return x 
```

```python
#resnet
import tensorflow as tf 
from tensorflow.keras import layers,activations 
from abc import ABC 
```

```python
#残差快
class Residual(tf.keras.Model,ABC):
    #定义网络结构
    def __init__(self,num_channels,use_1x1conv=False,strides=1):
        super(Residual,self).__init__()
        #卷积层 
        self.conv1 = layers.Conv2D(num_channels,padding="same",kernel_size=3,strides=strides)
        #卷积层
        self.conv2 = layers.Conv2D(num_channels,kernel_size=3,padding="same")
        #是否使用1*1的卷积
        if use_1x1conv:
            self.conv3 = layers.Conv2D(num_channels,kernel_size=1,strides=strides)
        else:
            self.conv3 = None
        #BN层
        self.bn1 = layers.BatchNormalization()
        self.bn2 = layers.BatchNormalization()
    #定义前向传播过程
    def call(self,x):
        Y = activations.relu(self.bn1(self.conv1(x)))
        Y = self.bn2(self.conv2(Y))
        if self.conv3:
            x = self.conv3(x)
        Y += X
        return tf.keras.activations.relu(Y)
```

```python
#残差模块
class ResnetBlock(tf.keras.layers.Layer):
    #定义网络结构
    def __init__(self,num_channels,num_res,first_block=False):
        super(ResnetBlock,self).__init__()
        #存储残差块
        self.listLayers = []
        #遍历残差快生成模块
        for i in range(num_res):
            #如果第一个残差块，而且不是第一个模块时
            if i == 0 and not first_block:
                self.listLayers.append(Residual(num_channels,use_1x1conv=True,strides=2))
            else:
                self.listLayers.append(Residual(num_channels))
    #前向传播
    def call_(self,x):
        for layer in self.listLayers:
            x = layer.call_(x)
        return x 
```

```python
#构建resnet网络
class ResNet(tf.keras.Model):
    #网络结构
    def __init__(self,num_blocks):
        super(ResNet,self).__init__()
        #输入层
        self.conv = layers.Conv2D(64,kernel_size=7,strides=2,padding="same")
        #BN层
        self.bn = layers.BatchNormalization()
        #激活层
        self.relu = layers.Activation('relu')
        #池化
        self.mp = layers.MaxPool2D(pool_size=3,strides=2,padding="same")
        #残差模块
        self.res_block1 = ResnetBlock(64,num_blocks[0],first_block=True,)
        self.res_block2 = ResnetBlock(128,num_blocks[1])
        self.res_block3 = ResnetBlock(256,num_blocks[2])
        self.res_block4 = ResnetBlock(512,num_blocks[3])
        self.gap = layers.GlobalAveragePooling2D()
        #全连接层
        self.fc = layers.Dense(units=10,activation=tf.keras.activations.softmax)
    #定义前向传播过程
    def call(self,x):
        #输入部分的传输过程
        x = self.conv(x)
        x = self.bn(x)
        x = self.relu(x)
        x = self.mp(x)
        #block
        x = self.res_block1(x)
        x = self.res_block2(x)
        x = self.res_block3(x)
        x = self.res_block4(x)
        #输出部分的传输
        x = self.gap(x)
        x = self.fc(x)
        return x 
```

```python
#finetuning
import tensorflow as tf 
import pathlib
import matplotlib.pyplot as plt
```

```python
#获取数据集
traindir = 'hotdog/train'
testdir = 'hotdog/test'
```

```python
image_gen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1/255.)
#获取训练集数据
train_data_gen = image_gen.flow_from_directory(traindir,batch_size=32,target_size=(224,224),shuffle=True)
```

```python
#获取测试机数据
test_data_gen = image_gen.flow_from_directory(testdir,batch_size=32,target_size=(224,224),shuffle=True)
```

```python
image,label = next(train_data_gen)
```

```python
plt.figure(figsize=(10,10))
for n in range(15):
    ax = plt.subplot(5,5,n + 1)
    plt.imshow(image[n])
    plt.axis('off')
```

```python
#模型构建与训练
ResNet50 = tf.keras.applications.ResNet50(weights='imagenet',input_shape=(224,224,3))
for layer in ResNet50.layers:
    layer.trainable = False
net  = tf.keras.models.Sequential()
net.add(ResNet50)
net.add(tf.keras.layers.Flatten())
net.add(tf.keras.layers.Dense(2,activation="sigmoid"))
```

```python
#模型编译
net.compile(optimizer='adam',loss=tf.keras.losses.binary_crossentropy,metrics=['accuracy'])
```

```python
#模型训练
net.fit(train_data_gen,steps_per_epoch=10,epochs=3,batch_size=32,validation_data=test_data_gen,validation_steps=10)
```

```python
#1目标检测
#iou
import numpy as np 
```

```python
#定义方法计算iou
def IOU(box1,box2,wh=False):
    #判断表示方式
    if wh == False:
        #极坐标
        xmin1,ymin1,xmax1,ymax1 = box1 
        xmin2,ymin2,xmax2,ymax2 = box2 
    else:
        #中心点坐标表示
        #第一个宽
        xmin1,ymin1 = int(box1[0]-box1[2]/2.0),int(box1[1]-box1[3]/2.0)
        xmax1,ymax1 =  int(box1[0]+box1[2]/2.0),int(box1[1]+box1[3]/2.0)
        #第二个宽
        xmin2,ymin2 = int(box2[0]-box2[2]/2.0),int(box2[1]-box2[3]/2.0)
        xmax2,ymax2 =  int(box2[0]+box2[2]/2.0),int(box2[1]+box2[3]/2.0)
    #获取交际的左上角和右下角
    xx1 = np.max(xmin1,xmin2)
    yy1 = np.max(ymin1,ymin2)
    xx2 = np.min(xmax2,ymax1)
    yy2 = np.min(ymax1,ymax2)
    #计算交际面积
    inter_area = (np.max(0,xx2-xx1)) *(np.max(0,yy2-yy1))
    #计算并的面积
    area1 = (xmax1-xmin1) *(ymax1-ymin1)
    area2 = (xmax2-xmin2)*(ymax2-ymin2)
    uion_area = area1 + area2 - inter_area
    #IOU
    IOU = inter_area / (uion_area+1e-6)
    return IOU 
```
